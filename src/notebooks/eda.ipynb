## What are the characteristics of popular songs?

[Spotify's Audio Feature](https://developer.spotify.com/documentation/web-api/reference/#/operations/get-several-audio-features)

Answer: 
* Majority track is not so acoustic, but rather danceable, energetic (i.e. fast, loud, and noisy)
* The tracks are mostly studio-production (not live), with less vocal (low instrumentalness) and not often feature speech sections
* The tracks are between 150,000-250,000 milliseconds or 2.5-4.2 mins, tempo ranges between 75 to 175 bpm
* Key-wise, more tracks are in C (0) or A-shape (10)
* Meanwhile, valence (or positiveness) varies across 0-1 scale but more song in the middle range 0.5
* There is also no preference between Major (1) and Minor (0) Mode

## Which tracks apear in multiple playlist?

Track that are features in multiple playlist maybe in 2 New Music Friday playlist, Rock classics, alternative that fits in multiple playlists, not necessarily the most popular hits.

```python
# t = x.groupby(by = ['track_name', 'artist_name']).get_group(('Wild Child', 'The Black Keys'))
```

```python
a = pd.DataFrame(x.groupby(by = ['track_name', 'artist_name']).apply(lambda x: list(x['playlist_name'])))
a = a.rename(columns= {0: "playlists"})
```

```python
b = pd.DataFrame(x.groupby(by = ['track_name', 'artist_name']).count()['track_uri'])
```

```python
c = x.groupby(by = ['track_name', 'artist_name']).mean()
```

```python
merge_df = a.merge(b, left_index=True, right_index=True).merge(c, left_index=True, right_index=True).query('track_uri >3')#.sort_values(by = 'track_uri', ascending=False).head(40)
```

```python
# distribution of song that features in at least 4 playlists among the 50 playlists
merge_df.hist(figsize=(15,12));
```
